```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IRIS.csv')
print("✅ Dataset Loaded Successfully!")
print(df.head())
```

```python
# User inputs (you can change these values)
x_edges = [4, 5, 6, 7, 8]     # edges of histogram bins
bar_color = 'skyblue'          # bar color
edge_color = 'black'           # edge color
alpha_value = 0.7              # transparency
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['sepal_length'], bins=x_edges, color=bar_color, edgecolor=edge_color, alpha=alpha_value)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='YlGnBu')
plt.title("Heatmap of Feature Correlations (Iris Dataset)")
plt.show()
```
