```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ---- Generate random ages ----
np.random.seed(42)  # for reproducibility
ages = np.random.randint(18, 70, size=100)  # ages between 18 and 70
```

```python
# ---- Display ages like barcode ----
plt.figure(figsize=(12, 4))
plt.bar(range(len(ages)), ages, color='black', width=1.0)

plt.title("Barcode-style Display of Ages for 100 Persons", fontsize=14)
plt.xlabel("Person Index")
plt.ylabel("Age")
plt.grid(False)  # No grid to make it look like a barcode
plt.show()

# Optional: Display first 10 generated ages
print("Sample of generated ages:", ages[:10])
```
